```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Berlin_crimes.csv')
```

```python
base = df.loc[df['Year'] == 2012].sum()
for year in df.Year.unique()[1:]:
    base = pd.concat([base, df.loc[df['Year'] == year].sum()], axis =1)
base.columns = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
df2 = base[4:]
```

```python
df2 = df2.transpose()
df2
```

```python
# Very unclear to see which crime is growing in this graph, better make a sort of delta table for df2
df2.plot(figsize=(15,7))
```

```python
# This is also pretty unclear, let's just look at if it rose or shrunk in the last three years
df3 = df2.iloc[:, 1] - df2.iloc[:, 0]
for i in range(2,len(df2.transpose())):
        df3 = pd.concat([df3, df2.iloc[:, i] - df2.iloc[:, i-1]], axis = 1)

df3.columns = ['2013', '2014', '2015', '2016', '2017', '2018', '2019']
df3
```
