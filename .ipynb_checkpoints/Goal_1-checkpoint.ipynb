```python
import pandas as pd
```

```python
df = pd.read_csv('Berlin_crimes.csv')
df
```

```python
# most crimes are committed in Mitte
base = df.loc[df['District'] == 'Mitte'].sum()
for dis in df.District.unique()[1:]:
    base = pd.concat([base, df.loc[df['District'] == dis].sum()], axis =1)
#base.columns = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
CrimeCum = base.drop(['Year','Code','Location', 'District'])
CrimeCum.columns = ['Mitte', 'Friedrichshain-Kreuzberg', 'Pankow', 'Charlottenburg-Wilmersdorf', 'Spandau', 
                   'Steglitz-Zehlendorf', 'Tempelhof-Schöneberg', 'Neukölln', 'Treptow-Köpenick', 'Marzahn-Hellersdorf',
                   'Lichtenberg', 'Reinickendorf']
CrimeCum = CrimeCum.sum()
CrimeCum
```

```python
# Inhabitants source: https://de.wikipedia.org/wiki/Verwaltungsgliederung_Berlins
# Mitte has the highest crime rate at 2367 crimes per 10.000 inhabitants on average per year note that this is also a very touristic area
inhab = {'Mitte' : 384172,
        'Friedrichshain-Kreuzberg' : 289762,
        'Pankow' : 407765,
        'Charlottenburg-Wilmersdorf' : 342332,
        'Spandau' : 243997,
        'Steglitz-Zehlendorf' : 308687,
        'Tempelhof-Schöneberg' : 351644,
        'Neukölln' : 329691,
        'Treptow-Köpenick' : 271153,
        'Marzahn-Hellersdorf' : 268548,
        'Lichtenberg' : 291452,
        'Reinickendorf' : 265225}

[[CrimeCum.keys()[i], int(10000/8 * CrimeCum[i] / inhab[CrimeCum.keys()[i]])] for i in range(len(CrimeCum))]
```
